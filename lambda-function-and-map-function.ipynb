```python
# Use of Map function

def attendance(attendance):                    # defining attendance function 
    if attendance>80:                    # with (variable attendance to store the value)
        return"good"
    elif attendance>60: 
        return"ok"
    else:
        return"improve it"
```

```python
attendance(29)
```

```python
std_marks=[28,45,59,74,96]
for i in std_marks:
    print(attendance(i))
```

```python
# using map function stored as list
std_marks1=list(map(attendance,std_marks))                       
std_marks1
```

```python
# using map function stored as dictionary
dict={}
for i in std_marks:
    for j in std_marks1:
        dict[i]=j
print(dict)
```
